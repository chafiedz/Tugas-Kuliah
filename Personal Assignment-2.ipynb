```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
ds = pd.read_csv('doctors_combined_data.csv')
```

## Data Exploration


```python
print("\nInfo dataset:")
# 5 data pertama
print(ds.head())
# statistik deskriptif
print("\nStatistik numerik:")
print(ds.describe().round(2))
# statistik deskriptif
print("Statistik deskriptif dasar:\n")

numeric_ds = ds.select_dtypes(include=np.number)

# Buat DataFrame ringkasan
statistik = pd.DataFrame({
    'Mean': numeric_ds.mean().round(2),
    'Median': numeric_ds.median().round(2),
    'Std Dev': numeric_ds.std().round(2),
    'Skewness': numeric_ds.skew().round(2),
    'Kurtosis': numeric_ds.kurtosis().round(2)
})

print(statistik)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(ds['Experience'], bins=25, kde=True, color='skyblue')
plt.title('Distribusi Experience')
plt.xlabel('Experience')
plt.ylabel('Frekuensi')
plt.show()

plt.figure(figsize=(6,5))
sns.boxplot(x=ds['Experience'], color='orange')
plt.title('Boxplot Experience')
plt.xlabel('Experience')
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(x='Experience', y='Salary', data=ds, color='green')
plt.title('Korelasi antara Experience dan Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(numeric_ds.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()



```
